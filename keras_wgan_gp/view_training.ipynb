```python
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
IMGS_FOLDER = 'imgs'

get_n = lambda x: int(x.split('_')[-1][:-4])
files = [os.path.join(IMGS_FOLDER, f) for f in os.listdir(IMGS_FOLDER) if f[-4:] == '.png']
files = sorted(files, key=get_n)


from IPython.display import HTML
from matplotlib import animation, rc

fig, ax = plt.subplots()

def animate(i):
    img = plt.imread(files[i])
    ax.imshow(img, interpolation='nearest')
```

```python
anim = animation.FuncAnimation(fig, animate, frames=len(files), interval=400)
HTML(anim.to_html5_video())
```
