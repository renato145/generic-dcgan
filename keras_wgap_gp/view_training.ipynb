```python
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
files = [os.path.join('imgs', f) for f in os.listdir('imgs/') if f[-4:] == '.png']

from IPython.display import HTML
from matplotlib import animation, rc

fig, ax = plt.subplots()

def animate(i):
    img = plt.imread(files[i])
    ax.imshow(img, interpolation='nearest')
```

```python
anim = animation.FuncAnimation(fig, animate, frames=len(files), interval=250)
HTML(anim.to_html5_video())
```
